```python
from pathlib import Path
import pandas as pd
import re
import os
```

```python
ROOT = Path.cwd().parent      # repo root when notebook is in /notebooks
DATA_PATH = ROOT / "data" / "basic_data"
OUT_PATH = ROOT / "data" / "processed"
OUT_PATH.mkdir(exist_ok=True)

df = pd.read_excel(DATA_PATH / "All_Circles_2023Q1_price.xlsx")
```

```python
print("Working directory:", os.getcwd())
print("File exists:", (DATA_PATH / "All_Circles_2023Q1_price.xlsx").exists())

df['Price2'] = df['Price'].str.extract(r'(\d+)').astype(float)
```

```python
def extract_days(x):
    match = re.search(r'(\d+)', str(x))
    return int(match.group(1)) if match else 'N.A.'

df['Days2'] = df['Validity'].apply(extract_days)


df[['Tag1', 'Tag2']] = df['Tags'].str.extract(r'(\w+)\s+(.*)')
```

```python
def compute_price_per_day(row):
    if isinstance(row['Days2'], int) and row['Days2'] > 0:
        return round(row['Price2'] / row['Days2'], 9)
    else:
        return None

df['clean_price_per_day'] = df.apply(compute_price_per_day, axis=1)

# 保存结果为 CSV 文件

df.to_csv(
    OUT_PATH / "processed_price_data_2023Q1.csv",
    index=False
)
```
