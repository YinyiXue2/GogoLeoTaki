```python
import pandas as pd
import re
```

```python
df = pd.read_excel(
    "/home/jupyter-xueyin/GogoLeoTaki/data/basic_data/All_Circles_2023Q1_price.xlsx"
)


df['Price2'] = df['Price'].str.extract(r'(\d+)').astype(float)
```

```python
def extract_days(x):
    match = re.search(r'(\d+)', str(x))
    return int(match.group(1)) if match else 'N.A.'

df['Days2'] = df['Validity'].apply(extract_days)


df[['Tag1', 'Tag2']] = df['Tags'].str.extract(r'(\w+)\s+(.*)')
```

```python
def compute_price_per_day(row):
    if isinstance(row['Days2'], int) and row['Days2'] > 0:
        return round(row['Price2'] / row['Days2'], 9)
    else:
        return None

df['clean_price_per_day'] = df.apply(compute_price_per_day, axis=1)

# 保存结果为 CSV 文件
df.to_csv(
    "/home/jupyter-xueyin/GogoLeoTaki/data/processed/processed_price_data_2023Q1.csv",
    index=False
)

```
